```python
import pandas as pd
import matplotlib.pyplot as plt

movies = pd.read_csv("https://raw.githubusercontent.com/wizardMati/expotraintelos/master/tmdb_movies.csv")
genres = pd.read_csv("https://raw.githubusercontent.com/wizardMati/expotraintelos/master/tmdb_genres.csv")


q3 = movies["vote_count"].quantile(0.75)

top_10_movies = (
    movies[movies["vote_count"] > q3]
    .sort_values("vote_average", ascending=False)
    .head(10)
)

print("\nTOP 10 najwyżej ocenianych filmów:")
print(top_10_movies[["title", "vote_average", "vote_count"]])

movies["release_date"] = pd.to_datetime(movies["release_date"], errors="coerce")
movies["year"] = movies["release_date"].dt.year

movies_2010_2016 = movies[
    (movies["year"] >= 2010) & (movies["year"] <= 2016)
]

movies_2010_2016 = movies_2010_2016[
    (movies_2010_2016["revenue"] > 0) &
    (movies_2010_2016["budget"] > 0)
]

yearly_stats = (
    movies_2010_2016
    .groupby("year")[["revenue", "budget"]]
    .mean()
    .reset_index()
)

plt.figure(figsize=(10, 6))
plt.bar(yearly_stats["year"], yearly_stats["revenue"], label="Średni przychód")
plt.plot(
    yearly_stats["year"],
    yearly_stats["budget"],
    marker="o",
    label="Średni budżet"
)

plt.xlabel("Rok")
plt.ylabel("Kwota (USD)")
plt.title("Średni przychód i budżet filmów (2010–2016)")
plt.ticklabel_format(style="plain", axis="y")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()

movies["genre_ids"] = movies["genre_ids"].str.strip("[]").str.split(",")

movies_exploded = movies.explode("genre_ids")
movies_exploded["genre_ids"] = movies_exploded["genre_ids"].astype(int)

movies_genres = movies_exploded.merge(
    genres,
    left_on="genre_ids",
    right_on="id",
    how="left"
)

genre_counts = movies_genres["name"].value_counts()

most_common_genre = genre_counts.idxmax()
most_common_count = genre_counts.max()

print("\nNajczęściej występujący gatunek:")
print(f"Gatunek: {most_common_genre}")
print(f"Liczba filmów: {most_common_count}")
avg_runtime_by_genre = (
    movies_genres
    .groupby("name")["runtime"]
    .mean()
    .sort_values(ascending=False)
)

longest_runtime_genre = avg_runtime_by_genre.idxmax()
longest_runtime_value = avg_runtime_by_genre.max()

print("\nGatunek o najdłuższym średnim czasie trwania:")
print(f"Gatunek: {longest_runtime_genre}")
print(f"Średni czas trwania: {longest_runtime_value:.2f} min")
runtime_data = movies_genres[
    movies_genres["name"] == longest_runtime_genre
]["runtime"].dropna()

plt.figure(figsize=(8, 5))
plt.hist(runtime_data, bins=20)
plt.title(f"Histogram czasu trwania – {longest_runtime_genre}")
plt.xlabel("Czas trwania (minuty)")
plt.ylabel("Liczba filmów")
plt.tight_layout()
plt.show()

```
